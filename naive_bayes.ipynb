```python
import pandas as pd
data=pd.read_csv('spam.csv',encoding='ISO-8859-1')
#puskal - here i use iso encode since i get encoding error while direct using csv file --------UnicodeDecodeError: 'utf-8' codec can't decode bytes in position 135-136: invalid continuation byte
```

```python
data.head()
```

```python
data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'], axis=1)
```

```python
data.groupby('v1').describe()
#this gives me rough idea about spam vs ham data in give dataset
```

```python
#this is for converting spam, ham string to numeric value ie. 1 & 0 respectively
data['spam']=data['v1'].apply(lambda x: 1 if x=='spam' else 0)
data.head()
```

```python
##puskal-- now i am going to import train test method using sklearn 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.v2, data.v1)

##we have to convert message into numberic i.e how many time in each message particular world occur ,we have to count that

##puskal here i use countvectoreizer but we can also use pipeline instead of this 
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:2]

#here x_train)count is message which is converted to numeric matrix as below output
```

```python
#puskal 
#naive bayes has Three classsification ile bernoulli, multinomial, gaussian 
#here i spam dataset has discrete data so i am going to use multinomial naive bayes
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
#testing our msg with accuracy

check_msg = [
    'puskal khadka is a sincere ai enthusist',
    'new year 2020 sale upto 60% exclusive discount'
    
]
check_msg_count = v.transform(check_msg)
model.predict(check_msg_count)
```

```python

X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)

#here i check accuracy
#step-1: convert x_test to count
#then transfer count to score method of model for prediction
```
